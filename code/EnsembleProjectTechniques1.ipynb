From the above performance on the training and testing data it is clearly evident that the model is performing well on the training data but not on the testing data. To increase the accuracy, pruning can be done.

```python
clf_pruned = DecisionTreeClassifier(criterion = "entropy", random_state = 100,
                               max_depth=3, min_samples_leaf=5)
clf_pruned.fit(X_train, y_train)
preds_pruned = clf_pruned.predict(X_test)
preds_pruned_train = clf_pruned.predict(X_train)
```

```python
from sklearn.metrics import accuracy_score,f1_score,recall_score,precision_score, confusion_matrix
print(accuracy_score(y_test,preds_pruned))
print(accuracy_score(y_train,preds_pruned_train))
```

From the above accuracy scores, it is evident that the accuracy score for the training data has decreased after pruning but the training data has an accuracy of only 79% and the accuracy of the testing data has increased to 78 % from 73%

```python
acc_DT = accuracy_score(y_test, preds_pruned)
```

True Positives (TP) - These are the correctly predicted positive values which implies that the value of actual class is True and the value of predicted class is also True.

True Negatives (TN) - These are the correctly predicted negative values which means that the value of actual class is false and value of predicted class is also false.

False Positives (FP) – When actual class is no and predicted class is yes. 

False Negatives (FN) – When actual class is yes but predicted class in no

precision- Precision is the ratio of correctly predicted positive observations to the total predicted positive observations. 

Precision = TP/TP+FP

Here, precision for 0 is 80%, 1 is 71%.

Here,it calculates the ratio of predicted positive normal class to the total predicted observations

Recall- Recall is the ratio of correctly predicted positive observations to the all observations in actual class - yes.

Recall= TP/TP+FN

Here, the recall for 0 class is 0.95, 1 is 35%
F1 score- F1 Score is the weighted average of Precision and Recall or harmonic mean of Precision and recall

F1score= 2*(Recall * Precision) / (Recall + Precision)

The F1score for 0 class is 87%
1 is 47%

True Positives (TP) - These are the correctly predicted positive values which implies that the value of actual class is True and the value of predicted class is also True.

True Negatives (TN) - These are the correctly predicted negative values which means that the value of actual class is false and value of predicted class is also false.

False Positives (FP) – When actual class is no and predicted class is yes. 

False Negatives (FN) – When actual class is yes but predicted class in no

precision- Precision is the ratio of correctly predicted positive observations to the total predicted positive observations. 

Precision = TP/TP+FP

Here, precision for 0 is 83%, 1 is 67%.

Here,it calculates the ratio of predicted positive normal class to the total predicted observations

Recall- Recall is the ratio of correctly predicted positive observations to the all observations in actual class - yes.

Recall= TP/TP+FN

Here, the recall for 0 class is 0.91, 1 is 50%
F1 score- F1 Score is the weighted average of Precision and Recall or harmonic mean of Precision and recall

F1score= 2*(Recall * Precision) / (Recall + Precision)

The F1score for 0 class is 87%
1 is 57%
